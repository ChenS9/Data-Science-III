```python
# Part 2
# 6
import pandas as pd 
import numpy as np
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt
import seaborn as sns
boston = load_boston()
df_boston = pd.DataFrame(boston.data, columns = boston.feature_names)
df_boston['MEDV'] = boston.target
# add the highPriced column
high_value_lists = []
for num in df_boston['MEDV'].tolist():
    if num > 40:
        high_value_lists.append(1)
    else:
        high_value_lists.append(0)
df_boston['highPriced'] = high_value_lists
# plot the variables      
sns.scatterplot(x="RM", y="MEDV",
              hue="highPriced",
              data=df_boston)

x_ = np.array([df_boston["RM"].min(), df_boston["RM"].max()])
y_ = -(-16.61067197/0.77286264) - (-1.85123297/0.77286264)*x_
plt.plot(x_, y_, '-r')
# From the plot we can see that there is a clear boundry between orange cluster and blue cluster.
# We can draw a line between orange cluster and the blue cluster when I split RM and MEDV by highPriced.
# So variable highPriced is linearly separable across the MEDV and RM.
```

```python
# 7
# store the values in a numpy.ndarray
df_boston_sub = df_boston[["RM","MEDV"]]
df_boston_target = df_boston[["highPriced"]]
X = df_boston_sub.values
Y = df_boston_target.values
# Shapes of X and Y
print(X.shape)
print(Y.shape)
```

```python
# 8 
# add a new column of ones
X = np.ones((506,3))
X[:,1:] = df_boston[["RM","MEDV"]].values
# print the first 5 rows of X
print(X[:5,:])
```

```python
# 9
# set the step size
start_stepsize = 0.0005
start_weights = np.random.rand(3).reshape(-1, 1)
start_stepsize,start_weights
```

```python
# 10
# build function myPerceptron
def myPerceptron(X, Y, weights, stepsize):
    # calculate the predicted y and updated weights
    Y_hat = np.where(np.dot(X, weights) >= 0, 1, 0)
    # calculate the error
    error = (Y - Y_hat)
    #count the sum of error
    count_mis = (error != 0).sum() 
    # calculate the updated weights
    delta_weights = np.dot(error.T, X).T
    updated_weights = weights + stepsize * delta_weights
    return updated_weights, count_mis
```

```python
%%time
# 11

cnt_mis_list = []
cnt_iter_list = []
# initialize
count_mis = 1
counter_iter = 0
stop_iter = 40000
# run the function until convergence
while count_mis > 0:
    if counter_iter == 0:
        updated_weights, count_mis = myPerceptron(X, Y, start_weights, start_stepsize)
    else:
        updated_weights, count_mis = myPerceptron(X, Y, updated_weights, start_stepsize)
    cnt_mis_list.append(count_mis)
    counter_iter += 1
    cnt_iter_list.append(counter_iter)
    if counter_iter >= stop_iter:
        print(f'Force quit for exceeding max iter: {stop_iter}')
        break
        
```

```python
# 11
# plot numbers of errors and interation number
plt.figure(figsize=(20, 10))
plt.plot(cnt_iter_list, cnt_mis_list)
plt.xlabel('Number of iterations')
plt.ylabel('Misclassification count')
plt.title('Misclassification count vs Number of iterations')
```

```python
# 12
# After converage, the weights are 
# W0 = -16.61067197, W1 = -1.85123297, W2 =0.77286264
# So the sepearation line is -16.61067197X1 - 1.85123297X2 + 0.77286264X3 = 0
# plot the separation line     
W0 = updated_weights[0][0]
W1 = updated_weights[1][0]
W2 = updated_weights[2][0]
# plot the variables and separation line    
sns.scatterplot(x="RM", y="MEDV",
              hue="highPriced",
              data=df_boston)

x_ = np.array([df_boston["RM"].min(), df_boston["RM"].max()])
y_ = -(W0/W2) - (W1/W2)*x_
plt.plot(x_, y_, '-r')

# From the plot, we can see that the separation line can split the data into two parts perfectly. And the results in
# last question also shows that the with the weights after converage, the misclassification reaches zero.
```

```python
# Part 3
# 14
from sklearn.model_selection import train_test_split
X,y = df_boston[df_boston.columns[0:13]],df_boston[df_boston.columns[13]]
X_train, X_test, y_train, y_test = train_test_split( 
    X, y, test_size=0.2)
df_train = df_all_cols = pd.concat([X_train, y_train], axis = 1)
df_test = df_all_cols = pd.concat([X_test, y_test], axis = 1)
print("The shape of training dataset is" ,df_train.shape)
print("The shape of test dataset is" ,df_test.shape)
```

```python
# 15
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import *
import numpy as np
# make an iteration list
iteration_list = [2,5,10,20,50]
np.random.seed(789)
for iteration in iteration_list:
    # build a single layer nn model
    nn_model = MLPRegressor(
hidden_layer_sizes = (iteration),max_iter=1000
)
    nn_model = nn_model.fit(X_train, y_train)
    # make prediction on training data and test data
    y_train_pred = nn_model.predict(X_train)
    y_test_pred = nn_model.predict(X_test)
    # calculate the mean squared error of training data
    mse_train = mean_squared_error(y_train, y_train_pred)
    # calculate the mean squared error of test data
    mse_test = mean_squared_error(y_test, y_test_pred)
    print("MSE of training data for size " + str(iteration) + " hidden layer is "+ str(mse_train))
    print("MSE of test data for size " + str(iteration) + " hidden layer is "+ str(mse_test))

# From the results, we can see that all the five models have higher performance on the training data, but the
# performance on the test data are lower. What's more, we can see that the model with 50 hidden layer size has
# the highest performace, mse achieving the lowest among all the 5 models. However, there is no trends show that
# the bigger the hidder layer sizes are, the higher the performance is. Obviously, the performance of 10 hidden
# layer size is lower than that of 5.
```

```python
# 16
# make an iteration list
iteration_lists = [(2,2),(5,5),(10,10),(20,20),(50,50)]
np.random.seed(567)
for size in iteration_lists:
    # build a two hidden layer nn model
    nn_multi_model = MLPRegressor(
hidden_layer_sizes = (size),max_iter=1000
)
    nn_multi_model = nn_multi_model.fit(X_train, y_train)
    # make prediction on training data and test data
    y_train_pred_m = nn_multi_model.predict(X_train)
    y_test_pred_m = nn_multi_model.predict(X_test)
    # calculate the mean squared error of training data
    mse_train_multi = mean_squared_error(y_train, y_train_pred_m)
    # calculate the mean squared error of test data
    mse_test_multi = mean_squared_error(y_test, y_test_pred_m)
    print("MSE of training data for size " + str(size) + " hidden layer is "+ str(mse_train_multi))
    print("MSE of test data for size " + str(size)+ " hidden layer is "+ str(mse_test_multi))

# From the results, we can see that all the five models have higher performance on the training data, but the
# performance on the test data are lower. The model with (10,10) size of hidden layer has the best performance on
# both training data and test data. The model with (2,2) size of hiden layer has the lowest performance.
```
